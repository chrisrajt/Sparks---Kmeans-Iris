```python
df = df.drop('Id', axis = 1)
```

```python
data= df.copy()
data = data.values
```

```python
dummies = pd.get_dummies(df['Species'], drop_first = True)
df = df.join(dummies)
df = df.drop('Species', axis = 1)
df
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 10), wcss)
plt.xlabel('Numbers')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(3)
kmeans.fit(df)
km_pred = kmeans.fit_predict(df)
km_pred
```

```python
sns.scatterplot(data[:,0], data[:,1],hue = km_pred, palette=['green','red', 'blue'])
```
